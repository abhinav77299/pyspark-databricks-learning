```python
my_ddl_schema = '''
                    Item_Identifier STRING,
                    Item_Weight STRING,
                    Item_Fat_Content STRING, 
                    Item_Visibility DOUBLE,
                    Item_Type STRING,
                    Item_MRP DOUBLE,
                    Outlet_Identifier STRING,
                    Outlet_Establishment_Year INT,
                    Outlet_Size STRING,
                    Outlet_Location_Type STRING, 
                    Outlet_Type STRING,
                    Item_Outlet_Sales DOUBLE 

                '''
```

```python
df = spark.read.format('csv')\
            .schema(my_ddl_schema)\
            .option('header',True)\
            .load('/Volumes/dbacademy/default/tutorials/BigMart Sales.csv')
```

```python
display(df)
```

```python
df.printSchema()
```

### Struct type()

```python
from pyspark.sql.types import * 
from pyspark.sql.functions import *
```

```python
df.select(col("Item_Weight"),col("Outlet_Establishment_Year"),col("Item_Fat_Content")).display()
```

```python
df.select(col("Item_MRP").alias("price")).display()
```

```python
df.display()
```

```python
df.filter(col("Item_Fat_Content")=="Regular").display()
```

```python
df.filter((col("Item_Type")=="Soft Drinks") & (col("Item_Weight")<10)).display()
```

```python
df.filter((col("Outlet_Size").isNull())&(col("Outlet_Location_Type")=="Tier 1")|(col("Outlet_Location_Type")=="Tier 2")).display()
```

```python
df.withColumnRenamed("Item_Fat_Content","FAT_CONTENT").display()
```

```python
df=df.withColumn("FLAG",lit("Y"))
```

```python
df.display()
```

```python
df=df.withColumn("Multiply",col("Item_MRP")*col("Item_Weight"))
```

```python
df.display()
```

```python
df=df.withColumn("Item_Weight",col("Item_Weight").cast("double"))
```

```python
df.printSchema()
```
