### Union Operations

```python
data1=[('1','Abhi'),('2','Ravi'),(None,'Ravi')]
df1=spark.createDataFrame(data1,['id','name'])

data2=[('12','Asi'),('2','Rhdjs'),(None,'Rhdi')]
df2=spark.createDataFrame(data2,['id','name'])
df1.display()
df2.display()
```

```python
df1.union(df2).display()
```

```python
data1 = [('kad','1',),
        ('sid','2',)]
schema1 = 'name STRING, id STRING' 

df1 = spark.createDataFrame(data1,schema1)

df1.display()
```

```python
df1.union(df2).display()
```

### Date Functions

```python
df=df.withColumn('CurrentDate',current_date())
df.display()
```

```python
df=df.withColumn('week_after',date_add('CurrentDate',7))
df.display()
```

```python
df=df.withColumn('week_after',date_add('CurrentDate',-13))
df.display()
```

```python
df=df.withColumn('Sub',date_sub('CurrentDate',7))
df.display()
```

```python
df = df.withColumn('datediff',datediff('CurrentDate','week_after'))
df.display()
```

### Collect List Example

```python
data = [('user1','book1'),
        ('user1','book2'),
        ('user2','book2'),
        ('user2','book4'),
        ('user3','book1')]

schema = 'user string, book string'

df_book = spark.createDataFrame(data,schema)

df_book.display()
```

```python
df_book.groupBy("user").agg(collect_list("book")).display()
```
